```python
def find_stat(lst):
    last_number = None
    if len(lst) >= 7:
        return "Effet"
    for i, el in enumerate(lst):
            try:
                int(el)
            except:
                pass
            else:
                last_number = i

    if lst[last_number+1] == "%":
        last_number += 1
    return ' '.join(lst[last_number+1:])
```

```python
def finder_number_of_digits(lst):
    num = 0
    for elt in lst:
        try:
            int(elt)
        except:
            pass
        else:
            num += 1
    return num
```

```python
def find_values_stat(lst):
    if '%' in lst:
        type = "percent"
    else:
        type = "flat"

    if len(lst) >= 7:
        type = "Effet"
        return {
            "type": type,
            "value": ' '.join(lst)
        }
    if finder_number_of_digits(lst) == 1:
        return {
            "type": type,
            "value": int(lst[0])
        }
    else:
        return {
            "type": type,
            "from": int(lst[0]) if int(lst[0]) < int(lst[2]) else int(lst[2]),
            "to": int(lst[2]) if int(lst[0]) < int(lst[2]) else int(lst[0])
        }
```
